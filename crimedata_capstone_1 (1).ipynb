## Extracting variables from the main dataframe

```python
crime=df[['DATE OCC','TIME OCC','AREA NAME','Rpt Dist No','Vict Age','Vict Sex','Vict Descent','Crm Cd Desc','Premis Desc','LAT','LON','LOCATION']]
crime.head(20)
```

## Cleaning

```python
#checking the missing value in crime dataframe
crime.isnull().sum()
```

```python
#changing the date occ column to datetime format using pd.to_datetime
crime['DATE OCC'] = pd.to_datetime(crime['DATE OCC'])
```

```python
crime['DATE OCC']
```

```python
crime['TIME OCC']
```

```python
crime['TIME OCC'] = pd.to_datetime(crime['TIME OCC'],format= '%H:%M' ).dt.time
```

```python
crime['TIME OCC']
```

```python
crime['TIME OCC']
```

```python
# Converting  the integer time to a string representation of time
crime['TIME OCC'] = crime['TIME OCC'].apply(lambda x: f'{x:04d}'[:2] + ':' + f'{x:04d}'[2:])

# Display the updated DataFrame
print(crime.head(10))
```

```python
# crime['TIME OCC'] = pd.to_datetime(crime['TIME OCC'], format = '%H:%M:%S').dt.time
crime['TIME OCC'] = pd.to_datetime(crime['TIME OCC'], format='%H:%M:%S').dt.strftime('%H:%M:%S')
# crime['TIME OCC'] = crime['TIME OCC'].astype('datetime64')
crime['TIME OCC']
```

```python
crime['TIME OCC']
```

```python
print('Rows:\t{}'.format(crime.shape[0]))
print('Variables:\t{}'.format(crime.shape[1]))
crime.dtypes
```

```python
#replacing space in the location column
crime['LOCATION'] = crime['LOCATION'].str.replace(r'\s+', ' ', regex=True)
```

```python
crime['LOCATION']
```

```python
#checking the percentage of missing value if it is below .5 we can remove the columns
(crime.isnull().sum()).sort_values(ascending=False) / crime.shape[0]
```

```python
#removing the missing value of vict descent,vict sex,premis desc because it is below .5 percentage
crime.dropna(subset=['Vict Descent', 'Vict Sex', 'Premis Desc'],inplace=True)
```

```python
#checking the missing value are correct or not
(crime.isnull().sum()).sort_values(ascending=False) / crime.shape[0]
```

```python
#removing the -1 and below number from the victim age
crime = crime[crime['Vict Age'] >= 1]
```

```python
#checking only column with "H" value from vict sex
crime_h = crime.loc[crime['Vict Sex'] == 'H']
crime_h
```

### Univariate Analysis:


```python
# Plot histograms for numerical columns
numerical_columns = crime.select_dtypes(include=['int64', 'float64']).columns
crime[numerical_columns].hist(bins=20, figsize=(15, 10))
plt.suptitle("Histograms for Numerical Columns")
plt.show()
```

**From this we can say that the most crime are occured in the first 700 district number
and from the vict age we can say that most of the victims are between the age of 25 to 40**

```python
# Plot bar plots for categorical columns
categorical_columns = ['AREA NAME','Vict Sex','Vict Descent']
plt.figure(figsize=(15, 20))

for i, column in enumerate(categorical_columns, 1):
    plt.subplot(4, 2, i)
    sns.countplot(data=crime, x=column, order=crime[column].value_counts().index)
    plt.title(f"Frequency of Categories in {column}")
    plt.xticks(rotation = 45)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the number of top categories to display
top_n = 20

# Plot bar plots for specific categorical columns
categorical_columns = ['Crm Cd Desc', 'Premis Desc','LOCATION']
plt.figure(figsize=(15, 20))

for i, column in enumerate(categorical_columns, 1):
    plt.subplot(4, 2, i)

    # Get the top n categories
    top_categories = crime[column].value_counts().nlargest(top_n).index

    # Filter the dataframe for only the top categories
    subset_data = crime[crime[column].isin(top_categories)]

    sns.countplot(data=subset_data, y=column, order=subset_data[column].value_counts().index)
    plt.title(f"Frequency of Categories in {column}")
    plt.xticks(rotation=48)

plt.tight_layout()
plt.show()
```

### bivariate analysis

```python
numerical_features = crime.select_dtypes(include='number')

# Construct the correlation matrix
correlation_matrix = numerical_features.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

In the heatmap, values closer to 1 indicate a strong positive correlation, while values closer to -1 indicate a strong negative correlation.

## Additional graphs

```python
#plot showing only the count of victim in the age group between 20 - 40

# Filter the DataFrame for victims aged between 20 and 40
filtered_crime = crime[(crime['Vict Age'] >= 20) & (crime['Vict Age'] <= 40)]

# Count the number of victims per sex within the age group
victim_count = filtered_crime['Vict Sex'].value_counts()

# Create a bar plot
sns.barplot(x=victim_count.index, y=victim_count.values)

# Set plot labels and title
plt.xlabel('Victim Sex')
plt.ylabel('Number of Victims')
plt.title('Number of Victims According to Sex within Age Group 20-40')

plt.show()
```

we can say from the above graph that females are more conc: in the age group of 20-40

```python

# Convert 'DATE OCC' to datetime format
#crime['DATE OCC'] = pd.to_datetime(crime['DATE OCC'])

# Extract the month from the 'DATE OCC' column
crime['Month'] = crime['DATE OCC'].dt.month

# Count the number of crimes per month
monthly_crime_counts = crime['Month'].value_counts().sort_index()

# Create a line plot
sns.lineplot(x=monthly_crime_counts.index, y=monthly_crime_counts.values, marker='o')

# Set plot labels and title
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.title('Monthly Crime Rates Over Time')

# Show the plot
plt.show()
```

as we can see that montly crime rates over time july till august show a increase in crime and also in january also show a high crime rate
